```python
class Solution:
    def uniquePaths(self, m: int, n: int) -> int:
        def countPaths(i, j, n, m):
            if i == (n - 1) and j == (m - 1):
                return 1
            if i >= n or j >= m:
                return 0
            else:
                return countPaths(i + 1, j, n, m) + countPaths(i, j + 1, n, m)


        return countPaths(0, 0, m, n)
```

```python
obj = Solution()
totalCount = obj.uniquePaths(3, 7)
print("The total number of Unique Paths are ", totalCount)
```

```python
# Here's a breakdown of the code:
# countPaths function:
# This is a recursive function that uses dynamic programming to calculate the number of unique paths. It takes the following parameters:
# i: The current row index.
# j: The current column index.
# n: The number of rows in the grid.
# m: The number of columns in the grid.
# dp: A 2D array to store the results of subproblems.
# Here's what the function does:
# Base cases:
# If i is equal to n-1 and j is equal to m-1, it means we've reached the bottom-right corner, so we return 1.
# If i is greater than or equal to n or j is greater than or equal to m, it means we've gone out of bounds, so we return 0.
# Memoization:
# If dp[i][j] is not equal to -1, it means we've already calculated the result for this subproblem, so we return the stored result.
# Recursive calls:
# Otherwise, we make two recursive calls: one for moving down (i+1) and one for moving right (j+1).
# We add the results of these two calls and store it in dp[i][j].
# Finally, we return the result.
# uniquePaths function:
# This function initializes the dp array and calls the countPaths function to calculate the number of unique paths.
```

```python
class Solution:
    def uniquePaths(self, m: int, n: int) -> int:
        N = n + m - 2
        r = m - 1
        res = 1
        for i in range(1, r + 1):
            res = res * (N - r + i) / i
        return int(res)
```

```python

# This is a solution to the "Unique Paths" problem, where you have an m x n grid and you need to find the number of unique paths from the top-left corner to the bottom-right corner. You can only move either down or right at any point in time.
# Here's a breakdown of the code:
# Mathematical Approach:
# The problem can be solved using combinatorics. The idea is to think of the problem as a sequence of m-1 downs and n-1 rights. We need to find the number of ways to arrange these downs and rights.
# This can be represented as a binomial coefficient:
# (m-1 + n-1)! / ((m-1)! * (n-1)!)
# Where ! denotes the factorial function.
# Code Explanation:
# The code implements this mathematical approach:
# Calculate N and r:
# N = n + m - 2: Calculates the total number of steps (m-1 downs and n-1 rights).
# r = m - 1: Calculates the number of downs.
# Initialize res:
# res = 1: Initializes the result variable to 1.
# Calculate the binomial coefficient:
# The loop calculates the binomial coefficient using the formula:
# (N! / (r! * (N-r)!) = (N * (N-1) * ... * (N-r+1)) / (r * (r-1) * ... * 1)
#    The loop iterates from `1` to `r` (inclusive), and at each iteration:


#     +   `res = res * (N - r + i) / i`: Updates the result by multiplying it with the current term and dividing it by `i`.
# Return the result:
# return int(res): Returns the final result as an integer.
# This solution has a time complexity of O(m) and a space complexity of O(1), making it efficient for large inputs.
```

```python
obj = Solution()
totalCount = obj.uniquePaths(2, 3)
print("The total number of Unique Paths are", totalCount)
```
