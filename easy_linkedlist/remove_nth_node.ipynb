```python
class Node:
    def __init__(self, data, next_node=None):
        self.data = data
        self.next = next_node
```

```python
arr = [1, 2, 3, 4, 5]
N = 3
head = Node(arr[0])
head.next = Node(arr[1])
head.next.next = Node(arr[2])
head.next.next.next = Node(arr[3])
head.next.next.next.next = Node(arr[4])
```

```python
def print_linked_list(head):
    temp = head
    while temp is not None:
        print(temp.data, end="-> ")
        temp = temp.next
    print()
```

```python
print_linked_list(head)
```

```python
# Function to delete the Nth node from the end of the linked list
def DeleteNthNodefromEnd(head, N):
    if head is None:
        return None
    cnt = 0
    temp = head

    # Count the number of nodes in the linked list or length
    while temp is not None:
        cnt += 1
        temp = temp.next

    # If N equals the total number of nodes, delete the head
    if cnt == N:
        newhead = head.next
        head = None
        return newhead

    # Calculate the position of the node to delete (res)
    res = cnt - N # length of linked list - nth from last = distance from head 
    print(res)
    temp = head

    # Traverse to the node just before the one to delete
    while temp is not None:
        res -= 1 # move towards  the node that would be deleted from head
        print(res)
        if res == 0: # reach the node that would be deleted
            break  # stop temp pointer previous node
        temp = temp.next

    # Delete the Nth node from the end
    delNode = temp.next # node that would be deleted
    temp.next = temp.next.next
    delNode = None # clean up memory
    return head
```

```python
head = DeleteNthNodefromEnd(head, N)
```

```python
print_linked_list(head)
```

Time Complexity: O(L)+O(L-N), We are calculating the length of the linked list and then iterating up to the (L-N)th node of the linked list, where L is the total length of the list.

Space Complexity:  O(1), as we have not used any extra space.

```python
# Function to delete the Nth node from the end of the linked list
def DeleteNthNodefromEnd(head, N):
    # Create two pointers, fastp and slowp
    fastp = head
    slowp = head

    # Move the fastp pointer N nodes ahead
    for i in range(N):
        fastp = fastp.next
        print(fastp.data)  # Print the data of the fastp node

    # If fastp becomes None, the Nth node from the end is the head
    if fastp is None:
        return head.next

    # Move both pointers until fastp reaches the end
    while fastp.next is not None:
        fastp = fastp.next
        print(fastp.data)  # Print the data of the slowp node
        slowp = slowp.next
        print(slowp.data)  # Print the data of the slowp node

    # Delete the Nth node from the end
    delNode = slowp.next
    slowp.next = slowp.next.next
    delNode = None
    return head
```

```python
head = DeleteNthNodefromEnd(head, N)
```

```python
print_linked_list(head)
```

Time Complexity: O(N) since the fast pointer will traverse the entire linked list, where N is the length of the linked list.

Space Complexity: O(1), as we have not used any extra space.
