```python
# Node class for the binary tree
class Node:
    def __init__(self, val):
        self.data = val
        self.left = None
        self.right = None
```

```python
root = Node(1)
root.left = Node(2)
root.right = Node(2)
root.left.left = Node(3)
root.right.right = Node(3)
root.left.right = Node(4)
root.right.left = Node(4)
```

```python
def print_tree(root):
    templ = None
    tempr = None
    if root.left is not None:
        print(f'rt->{root.data}, lc->{root.left.data}')
        templ = root.left
    if root.right is not None:
        print(f'rt->{root.data}, rc->{root.right.data}')
        tempr = root.right
    if templ is not None:
        print_tree(templ)
    if tempr is not None:
        print_tree(tempr)
```

```python
print_tree(root)
```

```python

class Solution:
    def is_symmetric_util(self, root1, root2):
        if root1 is None or root2 is None:
            return root1 == root2
        return (root1.data == root2.data) and self.is_symmetric_util(root1.left, root2.right) and self.is_symmetric_util(root1.right, root2.left)

    def is_symmetric(self, root):
        if root is None:
            return True
        return self.is_symmetric_util(root.left, root.right)
```

Depth-First Traversal (DFT)

```python
#     1
#    / \
#   2   3
#  / \   \
# 4   5   6


# The DFT traversal orders would be:
# Inorder: 4, 2, 5, 1, 3, 6
# Preorder: 1, 2, 4, 5, 3, 6
# Postorder: 4, 5, 2, 6, 3, 1
```

```python
def print_inorder(root):
    if root is None:
        return
    print_inorder(root.left)
    print(root.data, end=" ")
    print_inorder(root.right)
```

```python
solution = Solution()
res = solution.is_symmetric(root)
if res:
    print("This Tree is Symmetrical")
else:
    print("This Tree is NOT Symmetrical")
```

Time Complexity: O(N) where N is the number of nodes in the Binary Tree. This complexity arises from visiting each node exactly once during the traversal and the function compares the nodes in a symmetric manner.

Space Complexity: O(1) as no additional data structures or memory is allocated.

Breadth-First Traversal (BFT)

BFT is a traversal algorithm that visits all the nodes at the current level before moving to the next level.
Steps:
Start at the root node.
Visit all the nodes at the current level (i.e., the root node's children).
Move to the next level and visit all the nodes at that level.
Repeat steps 2-3 until all levels have been visited.

```python
#     1
#    / \
#   2   3
#  / \   \
# 4   5   6

# The BFT traversal order would be: 1, 2, 3, 4, 5, 6
```

```python
from collections import deque # ouble-ended queue
#  Elements are added and removed in a specific order (FIFO, LIFO, or a combination of both).
# Append (append()): Add an element to the end (back) of the deque.
# Prepend (appendleft()): Add an element to the beginning (front) of the deque.
# Pop (pop()): Remove an element from the end (back) of the deque.
# Popleft (popleft()): Remove an element from the beginning (front) of the deque.
```

```python
def breadth_first_traversal(root):
    if root is None:
        return

    queue = deque([root])
    while queue:
        node = queue.popleft()
        print(node.data, end=" ")

        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)
```

```python
print("Breadth-First Traversal:")
breadth_first_traversal(root)
```

 FIFO (First-In-First-Out) 

```python
def breadth_first_traversal2(root):
    if root is None:
        return

    queue = Queue()
    node = queue.put(root)
    while not queue.empty():
        node = queue.get()
        print(node.data, end=" ")

        if node.left:
            queue.put(node.left)
        if node.right:
            queue.put(node.right)
```

```python
print("Breadth-First Traversal:")
breadth_first_traversal2(root)
```
