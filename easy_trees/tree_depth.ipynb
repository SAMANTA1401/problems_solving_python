```python
# Node class for the binary tree
class Node:
    def __init__(self, val):
        self.data = val
        self.left = None
        self.right = None
```

```python
# Creating a sample binary tree
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.left.right.right = Node(6)
root.left.right.right.right = Node(7)
```

```python
def print_tree(root):
    templ = None
    tempr = None
    if root.left is not None:
        print(f'rt->{root.data}, lc->{root.left.data}')
        templ = root.left
    if root.right is not None:
        print(f'rt->{root.data}, rc->{root.right.data}')
        tempr = root.right
    if templ is not None:
        print_tree(templ)
    if tempr is not None:
        print_tree(tempr)
```

BFS

```python
class Solution:
    # Function to find the
    # maximum depth of a binary tree
    # using level order traversal
    def maxDepth(self, root):
        # If the root is NULL
        # (empty tree), depth is 0
        if root is None:
            return 0
        
        # Create a queue for
        # level order traversal
        q = Queue() # fifo first in first out        # stack lifo last in first out
        level = 0
        
        # Push the root node into the queue
        q.put(root) # Push the root node into the rear side 
        
        # While there are nodes in the queue
        while not q.empty():
            # Get the number of nodes
            # at the current level
            size = q.qsize()# left or right child or both children
            print(size)
            
            # add or Process all nodes
            # at the current level
            for i in range(size): # traverse all children(left, right)
                # Get the front node in the queue
                front = q.get() # remove the front node ( all children one by one )
                # print(front)
                
                # Enqueue left child if exists
                if front.left is not None:
                    q.put(front.left) # put the left child to rear 
                    
                
                # Enqueue right child if exists
                if front.right is not None:
                    q.put(front.right) # put the right child to rear
            
            # remove all children from current level
            # Increment level to
            # move to the next level
            level += 1
        
        # Return the level, which represents
        # the maximum depth of the tree
        return level
```

```python

solution = Solution()
depth = solution.maxDepth(root)
```

Time Complexity: O(N) where N is the number of nodes in the Binary Tree. This complexity arises from visiting each node exactly once during the traversal to determine the maximum depth.

Space Complexity: O(N) where N is the number of nodes in the Binary Tree because in the worst case scenario the tree is balanced and has N/2 nodes in its last level which will have to be stored in the queue.
